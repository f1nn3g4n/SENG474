# Problem 9: Guessing the source

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://gist.githubusercontent.com/inumanag/ebb1566746aba800899e406f03c799c1/raw/4dd9fab739281f0bdcbc0c1829fb1b4df2173446/prob9-data.txt',
                 header=None)
```

```python
x = np.array(df.index.tolist())
y = np.array(df[0].tolist())
plt.scatter(x, y)
```

```python
coeff = np.polyfit(x, y, 1)
fit = np.poly1d(coeff)
```

```python
plt.plot(x, y, "o", label="data")
plt.plot(x, fit(x), "--", label="fit")
plt.legend()
plt.show()
```

```python
# help from: https://stackoverflow.com/questions/29885408/how-plot-polyfit-n-log-n-with-python
# bit of trimming x, since x[0] = 0
coeff = np.polyfit(np.log(x[1:]), y[1:], 1)
fit = np.poly1d(coeff)
```

```python
plt.plot(x, y, "o", label="data")
plt.plot(x[1:], fit(np.log(x[1:])), "--", label="fit")
plt.legend()
plt.show()
```

Best feature transformation is np.log(x), since the data points clearly represent a log function.
